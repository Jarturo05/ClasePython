# NOTAS CLASE 1

Api Restfull --> Api basadas en Http es decir basadas en Get, post, put y delete
Get: Pedir u obtener la info de un producto
Post: Crear un nuevo producto
Put: Actualizar un producto
Delete: Eliminar un producto

JSON(Javascript Object Notation) es texto estructurado con llaves {} y pares clave-valor: El formato de los datos que trae al usuario, texto plano

Endpoint: Una URL especifica de tu API que hace algo, como http://miapi.com/prodctuso...etc

Rest(Representational State Transfer): Las reglas que lo ordenan

## Regla 1
Cada recurso tiene una URL única, cada espacio es un "endpoint"

## Regla 2
Usas verbos HTTP para indicar la acción.

Framework: Marco de trabajo establece normas, reglas y formas donde establece como se hacen las cosas.
